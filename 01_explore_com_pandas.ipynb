```python
df = df.rename(columns={
    "Ano de trabalho": "ano",
    "Nível de experiência": "senioridade",
    "Tipo de contratação": "contrato",
    "Cargo": "cargo",
    "Salário": "salario",
    "Moeda do salário": "moeda",
    "Salário em USD": "USD",
    "Residência do funcionário": "residencia",
    "Proporção de trabalho remoto": "remoto",
    "Localização da empresa": "empresa",
    "Tamanho da empresa": "tamanho_empresa"
})
```

```python
df["senioridade"] = df["senioridade"].replace({
    "SE": "Senior",
    "MI": "Pleno",
    "EN": "Junior",
    "EX": "Executivo"
})
df['senioridade'].value_counts()
```

```python
df["contrato"] = df["contrato"].replace({
    "FT": "Tempo integral",
    "PT": "Meio período",
    "CT": "Contrato temporário",
    "FL": "Freelancer"
})
df['contrato'].value_counts()
```

```python
df["tamanho_empresa"] = df["tamanho_empresa"].replace({
    "S": "Pequena",
    "M": "Média",
    "L": "Grande"
})
df['tamanho_empresa'].value_counts()
```

```python
df["remoto"] = df["remoto"].replace({
    0: "Presencial",
    50: "Híbrido",
    100: "Remoto"
})
df['remoto'].value_counts()
```
